```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries loaded successfully ✅")
```

```python
df = pd.read_csv('../data/advertising.csv', index_col=0)

print("Dataset loaded ✅")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].scatter(df['TV'], df['Sales'], color='steelblue', alpha=0.6)
axes[0].set_xlabel('TV Spend')
axes[0].set_ylabel('Sales')
axes[0].set_title('TV vs Sales')

axes[1].scatter(df['Radio'], df['Sales'], color='coral', alpha=0.6)
axes[1].set_xlabel('Radio Spend')
axes[1].set_ylabel('Sales')
axes[1].set_title('Radio vs Sales')

axes[2].scatter(df['Newspaper'], df['Sales'], color='green', alpha=0.6)
axes[2].set_xlabel('Newspaper Spend')
axes[2].set_ylabel('Sales')
axes[2].set_title('Newspaper vs Sales')

plt.tight_layout()
plt.savefig('../reports/spend_vs_sales.png')
plt.show()
print("Chart saved ✅")
```

```python
print("Correlation with Sales:")
print(df.corr()['Sales'].sort_values(ascending=False))
```
